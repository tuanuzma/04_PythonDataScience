### Outliers

```python
def percentile(data, percent):
    k = (len(data) - 1) * percent
    f = np.floor(k)
    c = np.ceil(k)
    if f == c:
        return data[int(k)]
    d0 = data[int(f)] * (c - k)
    d1 = data[int(c)] * (k - f)
    return d0 + d1
```

```python
sorted_data = sorted(population['pop_60'])
twentyfive = percentile(sorted_data, 0.25)
fifty = percentile(sorted_data, 0.5)
seventyfive = percentile(sorted_data, 0.75)
print(twentyfive, fifty, seventyfive)
```

```python
twentyfive = population['pop_60'].quantile(0.25)
fifty = population['pop_60'].quantile(0.5)
seventyfive = population['pop_60'].quantile(0.75)
print(twentyfive, fifty, seventyfive)
```

```python
iqr = seventyfive - twentyfive
iqr
```

```python
lowerfence = twentyfive - (1.5 * iqr)
lowerfence
```

```python
min(population['pop_60'])
```

```python
upperfence = seventyfive + (1.5 * iqr)
upperfence
```

```python
population[(population['pop_60'] < lowerfence) | (population['pop_60'] > upperfence)]
# the 2 rows are outliers
```

```python
population['pop'] == population['pop_18'] + population['pop_60'] + population['pop_12'] + population['pop_5']
# the categorical total is not same as the total column
```

```python
# pop_18 is also having the number that is pop_60
population['pop_18'] = population['pop_18'] - population['pop_60']
population
```

```python
# let us create a new column babies
population['pop_0'] = population['pop'] - (population['pop_18'] + population['pop_60'] + population['pop_12'] + population['pop_5'])
population
```

```python
population = population[['state', 'pop_0', 'pop_5', 'pop_12', 'pop_18', 'pop_60', 'pop']]
population
```

```python
population.drop('Total', axis=0, inplace=True, errors='ignore')
population.loc['Total'] = population.sum(numeric_only=True)
# .loc can accept 2 values row, column
population.loc['Total', 'state'] = 'Total'
population
```

### Conditional Selection

```python
population.loc[population['state'] != 'Total']
```

```python
# pull out the states where the number of babies is less than number of elderly
# also make sure you don't pull out the total row
population.loc[(population['babies'] < population['elderly']) & (population['state'] != 'Total')]
```

```python
population.loc[(population['babies'] + population['children'] < population['elderly']) & (population['state'] != 'Total')]
```

```python
# let us do some string based conditional selection
# Let us try some regular expressions
population.loc[population['state'].str.startswith('S')]
```

```python
population.loc[population['state'].str.startswith('W')]
```

```python
# pull out the state that starts with P and ends with g
# starts with can be written as ^P
# ends with can be written as g$
# in between P and g how many characters are allowed
# a single character is represented as .
# .+ one or more occurances of any character
# .* zero or more occurances of any character
population.loc[population['state'].str.contains('^P.+g$')]
```
