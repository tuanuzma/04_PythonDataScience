```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    'income':[3500, 2800, 3400, 4200, 2900],
    'expense':[2500, 1800, 2400, 3200, 1900]
}, index=['A', 'B', 'C', 'D', 'E'])
df
```

```python
df.plot(kind='line')
```

```python
# df.plot(kind='bar')
df.plot.bar()
```

```python
# df.plot(kind='barh')
df.plot.barh()
```

```python
df.plot(kind='pie', y='income')
```

```python
# df.plot(kind='pie', subplots=True)
df.plot.pie(subplots=True)
```
